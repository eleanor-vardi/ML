```python

%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
import pandas as pd
#!pip install folium
import folium
from folium.plugins import MarkerCluster
from folium.plugins import FastMarkerCluster
```

```python

df = pd.read_csv('flat_with_location.csv')
```

```python
loc_df = df[['latitude','longitude']]
loc_df = loc_df.dropna()
lats = loc_df['latitude'].tolist()
lons = loc_df['longitude'].tolist()
locations = list(zip(lats, lons))
```

```python
map1 = folium.Map(location=[45.5127, 4.4905], zoom_start=6)
FastMarkerCluster(data=locations).add_to(map1)
map1
```
